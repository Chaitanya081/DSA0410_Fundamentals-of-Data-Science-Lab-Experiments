```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("study_scores.csv")

print("Dataset Preview:")
print(df.head())
```

```python
# Select variables (CORRECT COLUMN NAMES)
study_time = df['Hours']
exam_score = df['Scores']
```

```python
# Select variables (CORRECT COLUMN NAMES)
study_time = df['Hours']
exam_score = df['Scores']
```

```python
# ---------------- SCATTER PLOT ----------------
plt.figure()
plt.scatter(study_time, exam_score)
plt.xlabel("Study Time (Hours)")
plt.ylabel("Exam Score")
plt.title("Scatter Plot: Study Time vs Exam Score")
plt.show()
```

```python
# ---------------- LINE PLOT ----------------
plt.figure()
plt.plot(study_time, exam_score, marker='o')
plt.xlabel("Study Time (Hours)")
plt.ylabel("Exam Score")
plt.title("Line Plot: Study Time vs Exam Score")
plt.show()
```

```python
# ---------------- REGRESSION PLOT ----------------
plt.figure()
sns.regplot(x=study_time, y=exam_score)
plt.xlabel("Study Time (Hours)")
plt.ylabel("Exam Score")
plt.title("Regression Plot: Study Time vs Exam Score")
plt.show()
```
