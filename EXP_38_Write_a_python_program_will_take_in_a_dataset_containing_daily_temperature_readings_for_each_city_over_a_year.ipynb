```python
import pandas as pd
```

```python
df = pd.read_csv("temperature.csv")

print("Dataset Preview:")
print(df.head())
```

```python
# Rename columns for clarity (if needed)
df.columns = ['Date', 'Temp']

# Calculate statistics
mean_temp = df['Temp'].mean()
std_temp = df['Temp'].std()
temp_range = df['Temp'].max() - df['Temp'].min()

print("\nTemperature Variability Analysis (Single City):")
print("Mean Temperature:", mean_temp)
print("Standard Deviation:", std_temp)
print("Temperature Range:", temp_range)
```

```python
if std_temp < 2:
    print("The city has a relatively consistent temperature.")
else:
    print("The city shows noticeable temperature variability.")
```
