```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("student_score.csv")
print(df.head())
```

```python
student_scores=df[["Math","Science","English","History"]].to_numpy()
```

```python
average_scores = np.mean(student_scores, axis=0)
```

```python
subjects = ["Math", "Science", "English", "History"]
```

```python
print("Average Scores:")
for i in range(len(subjects)):
    print(subjects[i],":",average_scores[i])

print("\nHighest Average subject:",subjects[np.argmax(average_scores)])
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("sales_performance.csv")
print(df.head())
```

```python
year_data=df[df["date"].str.contains("2000")]
sales_data=year_data["Revenue"].to_numpy()
total_sales=np.sum(sales_data)
percentage_increase=((sales_data[3]-sales_data[0])/sales_data[0])*100
```

```python
print("Quarterly Sales Data:", sales_data)
print("Total Sales for the Year:", total_sales)
print("Percentage Increase from Q1 to Q4:", percentage_increase, "%")
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Fuel_Efficiency.csv")
print(df.head())
```

```python
fuel_efficiency=df["highway_mpg(Fuel Efficiency)"].to_numpy()
average_efficiency = np.mean(fuel_efficiency)
model1=fuel_efficiency[0]
model2=fuel_efficiency[1]
percentage_increase=((model2-model1)/model1)*100
```

```python
print("Average Fuel Efficiency is:",average_efficiency)
print("Percentage Increase in Fuel Efficiency is:",percentage_increase)
```
