```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("montly_sales.csv",encoding="latin1")
print(df.head())
```

```python
plt.plot(df['MONTH_ID'], df['SALES'])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.scatter(df['MONTH_ID'], df['SALES'])
plt.title("Monthly Sales Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.bar(df['MONTH_ID'], df['SALES'])
plt.title("Monthly Sales Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
