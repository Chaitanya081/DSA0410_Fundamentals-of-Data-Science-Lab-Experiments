```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stock_prices.csv")

# Extract closing prices
close_prices = df['AAPL.Close']
```

```python
# Calculate variability measures
mean_price = close_prices.mean()
variance_price = close_prices.var()
std_dev_price = close_prices.std()
price_range = close_prices.max() - close_prices.min()
```

```python
print("Stock Price Variability Analysis")
print("Mean Closing Price:", mean_price)
print("Variance:", variance_price)
print("Standard Deviation:", std_dev_price)
print("Price Range:", price_range)
```

```python
# Plot stock price movement
plt.figure()
plt.plot(close_prices)
plt.xlabel("Trading Days")
plt.ylabel("Closing Price")
plt.title("Stock Closing Price Movement Over Time")
plt.show()
```
