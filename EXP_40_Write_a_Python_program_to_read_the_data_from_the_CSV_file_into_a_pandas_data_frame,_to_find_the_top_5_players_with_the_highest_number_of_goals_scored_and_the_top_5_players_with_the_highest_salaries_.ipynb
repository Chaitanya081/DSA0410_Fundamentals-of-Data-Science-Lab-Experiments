```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "soccer_player.csv",
    header=None,
    names=['Name', 'Age', 'Position', 'Goals', 'Weekly_Salary']
)
print("Column names:", df.columns)
print(df.head())
```

```python
# Top 5 players by goals
top_goals = df.sort_values(by='Goals', ascending=False).head(5)
print("\nTop 5 Players with Highest Goals:")
print(top_goals[['Name', 'Goals']])
```

```python
# Top 5 players by salary
top_salary = df.sort_values(by='Weekly_Salary', ascending=False).head(5)
print("\nTop 5 Players with Highest Weekly Salary:")
print(top_salary[['Name', 'Weekly_Salary']])
```

```python
# Average age
average_age = df['Age'].mean()
print("\nAverage Age of Players:", average_age)
```

```python
# Players above average age
above_avg_age = df[df['Age'] > average_age]
print("\nPlayers Above Average Age:")
print(above_avg_age[['Name', 'Age']])
```

```python
# Bar chart for positions
position_counts = df['Position'].value_counts()

plt.figure()
position_counts.plot(kind='bar')
plt.xlabel("Position")
plt.ylabel("Number of Players")
plt.title("Distribution of Soccer Players by Position")
plt.show()
```
