```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("Mall.csv")
print("Dataset Columns")
print(df.columns)
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Display first few clustered customers
print("\nClustered Data:")
print(df.head())
```

```python
print("\nEnter new customer details:")
income = float(input("Annual Income (in k$): "))
spending_score = float(input("Spending Score (1-100): "))

# Predict cluster for new customer
new_customer = pd.DataFrame(
    [[income, spending_score]],
    columns=['Annual Income (k$)', 'Spending Score (1-100)']
)

cluster = kmeans.predict(new_customer)

print("\nThe new customer belongs to Customer Segment:", cluster[0])
```
